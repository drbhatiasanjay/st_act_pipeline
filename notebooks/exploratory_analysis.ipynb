# ST-ACT Exploratory Data Analysis
This Jupyter Notebook visualizes compressed 4D cellular volumes loaded dynamically using Zarr v3 and Blosc2.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os
sys.path.append(os.path.abspath('..'))
from src.data_loader import AnisotropicZarrLoader

# Initialize loader
loader = AnisotropicZarrLoader('../data/cell_tracking_volume.zarr')
vol = loader.load_timepoint_block(0)
print('Volume Shape:', vol.shape)

# Plot heatmap of middle slice
plt.figure(figsize=(10, 6))
sns.heatmap(vol[vol.shape[0] // 2], cmap='magma')
plt.title('Z-Slice Signal Intensity (Seaborn)')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.show()
```
